## 코틀린의 여러가지 클래스 활용 및 람다 함수 학습

```kotlin
open class Shape {
    open fun draw() {
        println("도형 그리기")
    }
}

open class Circle : Shape() {
    override fun draw() { 
        println("Shape을 상속 받아 : ")
        println("원 그리기") 
        }
}

class Rect: Shape() {
    override fun draw() {
        println("사각형 그리기")
    }
    fun info() = println("사각형이다")
}

class RoundCircle: Circle()    //내용이 없으면 {} 생략가능

class Ellipse: Circle()


val shapes: List<Shape> = listOf(Circle(), Rect(), RoundCircle())

for (s in shapes) {
    s.draw()
}

shapes.forEach{ it.draw() }    //단일 argument인 경우에 it으로 대체가능 (s in shapes -> it)

//shapes[1].info(). //오류 발생, Rect객체이지만 부모 Shape의 참조변수에 할당이 되서, Rect를 몰라요

println("사각형 검사: ${shapes[1] is Rect}")
if (shapes[1] is Rect) { //현재 객체가 Rect에서 생성이 되었다면
    val r = shapes[1] as Rect //부모 클래스의 객체 변수 Rect로 변환 (downcasting)
    r.info()
    println(r::class)
}
//when을 이용해서 ...

for (s in shapes) {
    when(s) {
        is Ellipse -> println("타원형 객체")
        is Circle -> println("원형 객체")
        is Rect -> println("사각형 객체")
        is Shape -> println("도형 객체")
        else -> println("내가 모르는 객체")
    }
}
```

### 추상 클래스 학습

```kotlin
abstract class Animal1 //추상 클래스는 객체를 생성할 수 없다, open이 없어도 상속이 가능하다
```

```kotlin
//val ani = Animal()
abstract class Animal2 { //추상 메소드가 포함된 클래스는 반드시 추상 클래스로 연결
    abstract fun bark()
    open fun run() = println("달립니다")
}
```

```kotlin
class Dog : Animal1() {

}

val dog = Dog()
```

```kotlin
class Cat : Animal2() {
    override fun bark() = println("야옹")
    override fun run() = println("야옹이가 달리는데")
    fun run(name: String) = println(name) //메서드 오버라이딩, argument 수와 데이터 타입이 명시적으로 다른 경우에 동일한 이름을 갖는 메소드 형식
}

val cat = Cat()
cat.bark()
cat.run()
cat.run("메롱")
```

### interface

```kotlin
interface Flyable {
    fun fly() //추상 메소드
    fun land() = println("착륙") //기본 메소드 (default)
}

class Duck: Flyable {
    override fun land() = println("오리가 앉다")
    override fun fly() = println("오리가 날다")
}
```

```kotlin
val duck = Duck()
duck.land()
duck.fly()
```

### 클래스 위임

```kotlin
interface Printer {
    fun print(msg: String) //추상 메소드
}

class ColorPrinter: Printer {
    override fun print(msg: String) = println("[Color] $msg")
}
```

```kotlin
class Logger1: Printer {
    override fun print(msg: String) = println("[BW] $msg")
    fun log(msg: String) {
        print(msg)
    }
}

val logger1 = Logger1()
logger1.log("시작")
```

```kotlin
val color_printer = ColorPrinter()
color_printer.print("시작")
```

```kotlin
class Logger2(private val printer: Printer): Printer by printer {
    fun log(msg: String) {
        println(msg)
    }
}

val logger2 = Logger2(color_printer)
logger2.log("시작")
```

```kotlin
val double : (Int) -> Int = { it * 2}
val d = double(4)
println(d)
```

### 람다 함수

```kotlin
fun multiplier(factor: Int): (Int) -> Int = { it * factor } //입력 값에 factor를 곱해서 반환하는 함수
val unkown = multiplier(3)(5)
println("함수 동작: ${unkown}")
```

```kotlin
val some2: (Int) -> Unit = { println("$it 입력")}
some2(20)
```

```kotlin
fun testFun(no: Int, arg: (Int) -> Boolean): () -> String {
    val result = if(arg(10)) { //(Int)->Boolean, true/false
        "valid"
    } else {
        "invalid"
    }

    return { "testFun result: $result" } //{ () -> "..." } => { -> "..." } => { "..." }
}

val testResult = testFun(10) { no -> no > 0 }
testResult()
```
